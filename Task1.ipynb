```python
df.drop(['Indicator Name','Country Name'],axis=1,inplace=True)
```

```python
#Filtering Data
total_population_data=df[df['Indicator Code']=='SP.POP.TOTL']

#Sort Data based on total population for 2023
total_population_sorted=total_population_data.sort_values(by='2023',ascending=False)

#Get the top ten countries with the highest total population for 2023
total_top_ten_countries=total_population_sorted.head(10)

print('Top ten countries of total population \n')
print(total_top_ten_countries[['Country Code']])
```

```python
#Bar PLot
#Top ten countries in population in 2023 and 2024
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='2022',y='Country Code',data=total_top_ten_countries,palette='coolwarm',hue='Country Code',legend=False)
plt.title('Top ten countries of total population(2022)',fontsize=10)
plt.xlabel('Total Population',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='2024',y='Country Code',data=total_top_ten_countries,palette='coolwarm',hue='Country Code',legend=False)
plt.title('Top ten countries of total population(2024)',fontsize=10)
plt.xlabel('Total Population',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.show()
```

```python
#Extraction of bottom 10 data
#Sort Data based on total population for 2023
total_population_sorted1=total_population_data.sort_values(by='2022',ascending=True)

#Get the bottom ten countries with the highest total population for 2023
total_bottom_ten_countries=total_population_sorted1.head(10)

print('Bottom ten countries of total population \n')
print(total_bottom_ten_countries[['Country Code']])
```

```python
#Bar PLot
#Bottom ten countries in population in 2023 and 2024
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='2022',y='Country Code',data=total_bottom_ten_countries,palette='coolwarm',hue='Country Code',legend=False)
plt.title('Bottom ten countries of total population(2022)',fontsize=10)
plt.xlabel('Total Population',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='2024',y='Country Code',data=total_bottom_ten_countries,palette='coolwarm',hue='Country Code',legend=False)
plt.title('Bottom ten countries of total population(2024)',fontsize=10)
plt.xlabel('Total Population',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.show()
```
